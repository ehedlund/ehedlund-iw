```python
import ast
import pandas as pd
```

```python
with open('resource_counts.txt', 'r') as f:
    s = f.read()
    resource_dict = ast.literal_eval(s)
```

```python
resource_counts = pd.DataFrame(list(resource_dict.items()), columns=['url', 'count'])
resource_counts
```

```python
resource_counts = resource_counts.sort_values(by='count', ascending=False)
resource_counts
```

```python
resource_counts = resource_counts[resource_counts['count'] != 1]
resource_counts
```

```python
resource_counts = resource_counts[resource_counts['count'] < 255]
ax = resource_counts.plot.bar(rot=0)
```

```python
with open('ad_results.txt', 'r') as f:
    s = f.read()
    site_dict = ast.literal_eval(s)
```

```python
site_counts = pd.DataFrame(list(site_dict.items()), columns=['url', 'count'])
site_counts = site_counts.sort_values(by='count', ascending=False)
site_counts
```

```python
site_counts.to_csv('resources_per_site.csv')
```
